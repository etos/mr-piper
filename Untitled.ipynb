```python
import os
import shutil
from executor import ExternalCommand
import re
import json

regex = "([^ ]+) \(([0-9.]+)\)"



class Nest(object):
    
    def __init__(self):
        self.dependencies = {
            'dependency-name': {
                'version': '',
                'dependencies': {}
            }
        }
    
    def setup(self):
        # create virtualenv
        self.create_lock()
        self.create_requirements_structure()
        self.create_virtual_env()
    
    def cleanup(self):
        pass
    
    # SETUP METHODS
    
    def create_lock(self):
        with open(".nest.lock", "wb") as file:
            json.dump({}, file, indent=4)

    def create_requirements_structure(self):
        shutil.rmtree("requirements")
    #     os.rmdir("requirements", recursive=True)
        os.mkdir("requirements")
        open("./requirements/local.txt", "a").close()
        open("./requirements/production.txt", "a").close()
        open("./requirements/base.txt", "a").close()
        open("./requirements/frozen.txt", "a").close()

    def create_virtual_env(self, name="python_env"):
        output = self.run("virtualenv {name}".format(name=name), envs={
            'WORKON_HOME':'.'
        })
        print output
        
    # END SETUP METHODS
    
    # CLEANUP METHODS

    def delete_virtual_env(self, name="python_env"):
        shutil.rmtree(name)
        
    # END CLEANUP METHODS
    
    def save_lock(self):
        with open(".nest.lock", "wb") as file:
            json.dump(self.dependencies, file, indent=4)
            
    def load_lock(self):
        with open(".nest.lock", "rb") as file:
            self.dependencies = json.load(self.dependencies, file)
        
    
    def run(self, command, input=None, virtualenv=None, envs={}):
        runner = ExternalCommand(command, capture=True, check=False, input=input, virtual_environment=virtualenv, environment=envs)
        runner.start()
        return runner.output
    
    def test_install_parallel(self):
        packages = ["fabric", "django"]
        

    def run_env(self, command):
        return self.run(command, virtualenv="python_env", input='y')

    # PUBLIC METHODS
    
    def add(self, package):
        # create temo env, install package, get deps, destroy temp env, install package in env
        
        self.delete_virtual_env("temp_env")
        self.create_virtual_env("temp_env")
        self.mark_installed_packages(virtualenv="temp_env")
        output1 = self.run("pip install {package}".format(package=package), virtualenv="temp_env", input='y')
        diff = self.get_installed_package_difference(virtualenv="temp_env")
        
        output2 = self.run("pip install {package}".format(package=package), virtualenv="python_env", input='y')
        
        print output1
        print diff
        print self.list_packages(virtualenv="temp_env")
        
        package_name = ""
        package_version = ""
        for name, version in diff["all"].iteritems():
            if name.lower().strip() == package:
                print "PACKAGE IS", name, version
                package_name = name
                package_version = version
        del diff["all"][package_name]
        
        self.dependencies[package_name] = {
            'version': package_version,
            'dependencies': diff["all"]
        }
        print self.dependencies
        self.save_lock()

    def detect_new_packages(self):
        pass

    def remove(self, package):
        output = self.run_env("pip uninstall {package}".format(package=package))
        print output

    def list_packages(self, virtualenv="python_env"):
        output = run("pip list --format=legacy", virtualenv=virtualenv)
        print output
        
    def mark_installed_packages(self, virtualenv="python_env"):
        self.installed_before = self.get_installed_packages(virtualenv=virtualenv)

    def get_installed_packages(self, virtualenv="python_env"):
        r = re.compile("^([^ ]+) \(([0-9.]+)\)", re.MULTILINE)
        installed = r.findall(run("pip list", virtualenv=virtualenv))
        return dict(installed)

    def get_installed_package_difference(self, virtualenv="python_env"):
        before = self.installed_before
        now = self.get_installed_packages(virtualenv=virtualenv)

        data = {
            'all': {},
            'new': {},
            'updated': {},
            'downgraded': {}
        }

        for name, version in now.iteritems():
            
            if name in before.keys():
                if version == before[name]:
                    # nothing changed
                    pass
                else:
                    # updated
                    data["all"][name] = version
                    data["updated"][name] = version
            else:
                # new package
                data["all"][name] = version
                data["new"][name] = version

        return data




    def init(self):
        pass

    
nest = Nest()
nest.setup()
nest.add("fabric")
# nest.delete_virtual_env()
# nest.create_virtual_env()
# nest.create_requirements_structure()
# nest.before = get_installed_packages()
# nest.list_packages()
# add("fabric")
# print nest.get_installed_package_difference(before)
# nest.list_packages()
# nest.remove("fabric")
# nest.list_packages()
    
```
